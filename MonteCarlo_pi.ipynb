```
import matplotlib.pyplot as plt
import random
```

```
N = 1000
pi_quarter = 0
plt.axes().set_aspect('equal')
for i in range(N):
    x = random.random()
    y = random.random()
    if (x**2 + y**2) < 1:
        pi_quarter += 1
        plt.scatter(x, y, c='blue')
    else:
        plt.scatter(x, y, c='green')
print(str(pi_quarter)+" / "+str(N))
print("predicted pi: "+str(pi_quarter*4/N))

plt.show()
```
